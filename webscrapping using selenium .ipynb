```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import sqlalchemy
```

```python
Bus_Links=[]
driver = webdriver.Chrome()  # Optional argument, if not specified will search path.
Bus1=driver.get('https://www.redbus.in/')
driver.maximize_window()
bb=driver.find_element(By.XPATH,"//*[contains(text(),'Can I book a Government bus ticket on redBus?')]")
bb.click()
cc=driver.find_element(By.XPATH,"//*[contains(text(),'Yes, you can book government bus tickets on redBus. redBus has extended its bus booking services to many RTCs in India. Some of these RTCs are ')]")
BusLinks=cc.find_elements(By.TAG_NAME,'a')
for i in BusLinks:
    Bus_Links.append(i.get_attribute("href"))
```

```python
Bus_routes_name=[]
Bus_routes_links=[]
driver = webdriver.Chrome()  # Optional argument, if not specified will search path.
for j in Bus_Links: 
    Buses=driver.get(j)
    driver.maximize_window()
    pages=driver.find_elements(By.XPATH,"//div[@class='DC_117_pageTabs ']")
    if len(pages)>0:
        for i in range(0,len(pages)+1):
            routes=driver.find_elements(By.CLASS_NAME,"route")
            for r in routes:
                Bus_routes_name.append(r.text)
            links=driver.find_elements(By.CLASS_NAME,"route")
            for l in links:
                Bus_routes_links.append(l.get_attribute("href"))
            if i==len(pages):
                i=len(pages)-1
            driver.execute_script("arguments[0].click()",pages[i])
```

```python
Bus_names=[]
Bus_routes_names=[]
Bus_route_link=[]
Bus_type=[]
Departing_time=[]
Duration=[]
Reaching_time=[]
rate=[]
Price=[]
Seat_available=[]
a=[]
p=[]
j=0
for i in Bus_routes_links:
    driver.get(i)
    prev_height = -1
    max_scrolls = 100
    scroll_count = 0
      # Optional argument, if not specified will search path.
    time.sleep(5) # wait for some 5 sec!
    button=driver.find_elements(By.XPATH,"//div[contains(text(),'View Buses')]")
    for b in button:
        driver.execute_script("arguments[0].click()",b)
    while scroll_count < max_scrolls:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)  # give some time for new results to load
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == prev_height:
            break
        prev_height = new_height
        scroll_count += 1
    Busnames=driver.find_elements(By.XPATH,"//div[@class='travels lh-24 f-bold d-color']")
    Bustype=driver.find_elements(By.XPATH,"//div[@class='bus-type f-12 m-top-16 l-color evBus']") 
    Departingtime=driver.find_elements(By.XPATH,"//div[@class='dp-time f-19 d-color f-bold']")
    Durations=driver.find_elements(By.XPATH,"//div[@class='dur l-color lh-24']")
    Reachingtime=driver.find_elements(By.XPATH,"//div[@class='bp-time f-19 d-color disp-Inline']")
    starrate=driver.find_elements(By.XPATH,"//div[@class='column-six p-right-10 w-10 fl']")
    Prices=driver.find_elements(By.XPATH,"//*[@class='f-bold f-19' or @class='f-19 f-bold']")
    Seatavailable=driver.find_elements(By.XPATH,"//div[contains(@class,'seat-left m-top-')]")
    Date_reach_container=driver.find_elements(By.XPATH,"//div[@class='column-five p-right-10 w-10 fl']")
    Date_start_container=driver.find_elements(By.XPATH,"//div[@class='column-three p-right-10 w-10 fl']")
    for bu in Busnames:
        Bus_names.append(bu.text)
        Bus_route_link.append(i)
        Bus_routes_names.append(Bus_routes_name[j])
    j=j+1
    for bt in Bustype:    
        Bus_type.append(bt.text)
    for de in Departingtime:
        Departing_time.append(de.text)
    for du in Durations:
        Duration.append(du.text)
    for re in Reachingtime:
        Reaching_time.append(re.text)
    for pr in Prices:
        Price.append(pr.text)
    for sa in Seatavailable:
        Seat_available.append(sa.text)
    for sr in starrate:
        rate.append(sr.text)
    for con in Date_reach_container:
        a.append(con.text)
    for con1 in Date_start_container:
        p.append(con1.text)
    
   
    
        
        
  
```

```python
sr=[]
StarRating=[]
for i in rate:
    sr.append(i.split("\n"))
for i in sr:
    if '.' in i[0]:
        StarRating.append(float(i[0]))
    else:
        StarRating.append("Null")
        
```

```python
r_dates=[]
from datetime import date, timedelta
b=[]
for value in a:
    b.append(value.split("\n"))
for val in b:
    if len(val)==2:
        r_dates.append(date.today())
    else:
        r_dates.append(date.today()+timedelta(1))
```

```python
s_dates=[]
from datetime import date, timedelta
q=[]
for value in p:
    q.append(value.split("\n"))
for val in q:
    if len(val)==2:
        s_dates.append(date.today())
    else:
        s_dates.append(date.today()+timedelta(1))
```

```python
from datetime import date,datetime
Departing_date_time=[]
Reaching_date_time=[]
for i in range(0,len(Reaching_time)):
    a=str(r_dates[i])
    string_time=a+" "+Reaching_time[i]+":00"
    time_object = datetime.strptime(string_time, '%Y-%m-%d %H:%M:%S')
    date_time=str(time_object)
    Reaching_date_time.append(date_time)
for i in range(0,len(Departing_time)):
    a=str(s_dates[i])
    string_time=a+" "+Departing_time[i]+":00"
    time_object = datetime.strptime(string_time, '%Y-%m-%d %H:%M:%S')
    date_time=str(time_object)
    Departing_date_time.append(date_time)
```

```python
seatsavailable=[]
for sa in Seat_available:
    li=(sa.split(" "))
    for char in li:
        if char.isdigit():
            seatsavailable.append(int(char))
```

```python
Bus_types=[]
for i in Bus_type:
    Bus_types.append(i.replace('A/C','AC').replace('A.C','AC').replace('Non AC','NON AC').replace('NON-AC','NON AC'))
```

```python
sn=[]
rn=[]
RoutesNames=[]
Starting_names=[]
Reaching_names=[]
for i in Bus_routes_names:
    RoutesNames.append(i.split(" "))
for i in RoutesNames:
    temp=i.index("to")
    sn.append(i[:temp])
    rn.append(i[temp+1:])
for i in sn:
    st_name=" ".join(i)
    Starting_names.append(st_name)
for i in rn:
    rch_name=" ".join(i)
    Reaching_names.append(rch_name)
```
